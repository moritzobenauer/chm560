```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, ax1 = plt.subplots(1,1, sharex=False, figsize=(8,6))

ax2 = ax1.twinx()

vals = []

for k in [16,32,48,64]:
    df = pd.read_csv(f'binder/{k}x{k}_0.0.csv')

    binder_cum = 1 - df['M4'].values / (3*df['M2'].values**2)
    j = df['J'].values

    ax1.plot(j, binder_cum, '-o', label=k)

    vals.append(binder_cum)

ax1.set_ylabel(r'Binder Cumulants $B$', fontsize=14)
ax1.set_xlabel(r'$J / kT$', fontsize=14)


data = np.vstack([vals[0], vals[1], vals[2], vals[3]])
std=np.std(data, axis=0)
print(np.argmin(std))
print(j[np.argmin(std)])

ax2.plot(j,std, color='k', lw=2, alpha=0.5)
ax2.axvline(j[np.argmin(std)], lw=2, linestyle='dashed', alpha=0.5, color='red')
ax2.set_ylabel(r'$\sigma(\Sigma_i B_i)$', fontsize=14)
ax2.set_xlabel(r'$J / kT$', fontsize=14)

ax1.legend()
```
